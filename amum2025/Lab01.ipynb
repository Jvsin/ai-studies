### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
# TUTAJ UMIESC KOD
def function(x):
    return (x**4)/8 + 2*((x + 0.7)*(x + 0.7)*(x + 0.7)) - (x**2)/2 - 38 * x + 1

function(0)
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
# TUTAJ UMIESC KOD
results = []

for i in range(-20, 20):
    results.append(function(i))

print(results)
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
# TUTAJ UMIESC KOD
import matplotlib.pyplot as plt

X = range(-20, 20)
Y = results

plt.figure(figsize=(5,5))
plt.title("Wykresik")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.plot(X,Y)
# plt.xlim(2.5, 3.5) ## pokazywanie fragmentu wykresu (zoom na x)
# plt.ylim(-50, 0) ## fragment wyrkresu po y
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
# TUTAJ UMIESC KOD
min_value = min(results)
min_index = results.index(min_value)
min_x = X[min_index]

print("Min: " + str(min_value) + " dla x=" + str(min_x))
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
# TUTAJ UMIESC KOD
new_results = []
start, stop, step = -20, 20, 0.9
float_range = [start + i * step for i in range(int((stop - start) / step))]
for i in float_range:
    new_results.append(function(i))
    
min_value = min(new_results)
min_index = new_results.index(min_value)
min_x = float_range[min_index]
print("Min: " + str(min_value) + " dla x=" + str(min_x))
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
# TUTAJ UMIESC KOD
X = float_range
Y = new_results

plt.figure(figsize=(5,5))
plt.title("Wykresik drugi dla delta " + str(step))
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.ylim(-50, 0)
plt.scatter(min_x, min_value, color='red', zorder=3, label=f"Minimum ({min_x}, {min_value:.2f})")
plt.annotate(f"({min_x}, {min_value:.2f})", (min_x, min_value), textcoords="offset points", xytext=(-30,10), ha='center', fontsize=10, color='red')
plt.plot(X,Y)
```

### Zad 7:

Wygeneruj $k$ losowych wartości $x$ w przedziale [-20, 20](metoda Monte Carlo), a następnie znajdź minimum funkcji $f(x)$.
Wyświetl wykres funkcji.

```python
# TUTAJ UMIESC KOD
import random

k = 1000 
x_random = sorted([random.uniform(-20, 20) for _ in range(k)])
y_random = [function(x) for x in x_random]

min_value = min(y_random)
min_index = y_random.index(min_value)
min_x = x_random[min_index]
print("Min: " + str(min_value) + " dla x=" + str(min_x))

plt.figure(figsize=(5,5))
plt.title("Wykresik trzeci")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.ylim(-50,-30)
plt.scatter(min_x, min_value, color='red', zorder=3, label=f"Minimum ({min_x}, {min_value:.2f})")
plt.annotate(f"({min_x}, {min_value:.2f})", (min_x, min_value), textcoords="offset points", xytext=(-30,10), ha='center', fontsize=10, color='red')
plt.plot(x_random, y_random)
```
